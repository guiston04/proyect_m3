```python
#Import Libraries

import pandas as pd
```

```python
#Random Forest results

df1 = pd.read_csv('./results_rf.csv')

# Deep Learning Results

df2 = pd.read_csv('./results_rn.csv')

# CatBooster results

df3 = pd.read_csv('./results_catb.csv')

# XGB results

df4 = pd.read_csv('./results_xg.csv')
```

```python
# Merge of all the results

prediction = df1.merge(df2, on = 'id')
prediction = prediction.merge(df3, on = 'id')
prediction = prediction.merge(df4, on = 'id')
```

```python
prediction
```

```python
# Weighted average of the results

prediction['Price_u'] = (5*prediction['Price_x']+5*prediction['Price_y']+40*prediction['Price_xg']+50*prediction['Price'])/100
```

```python
prediction
```

```python
# Formatting of the dataset for Kaggle

prediction.drop('Price_x',inplace = True, axis = 1)
prediction.drop('Price_y',inplace = True, axis = 1)
prediction.drop('Price',inplace = True, axis = 1)
prediction.drop('Price_xg',inplace = True, axis = 1)
prediction = prediction.rename(columns={'Price_u': 'Price'})
prediction = pd.DataFrame(prediction, columns = ['Price'])
prediction.index.names = ['id']
```

```python
prediction
```

```python
# Creation of file where predictions are saved

prediction.to_csv('./results.csv')
```
